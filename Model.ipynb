```python
#Filling NaN entries in the 'days_since_prior_order' column

MEAN = orders['days_since_prior_order'].mean()
orders['days_since_prior_order'].fillna(MEAN,inplace = True)
sum(orders['days_since_prior_order'].isna())
```

```python
present = np.zeros(206209)

for i in range(316717):
    UserID = int(orders.loc[i]['user_id'])
    present[UserID - 1] += 1

present
```

```python
import seaborn as sns

sns.countplot(x = orders['order_hour_of_day'],data = orders)
```

```python
#People tend to place more orders int the hours b/w 9-16

def ActiveTime(UserID):
    
    # Users who order at night between 0-5 and 21-24 hours are returned 1
    # Users who order in morning between 5-13 hours are returned 2
    # Users who order in evening between 13-21 hours are returned 3
    
    MeanTime = orders.loc[orders['user_id'] == UserID]['order_hour_of_day'].mean()
    
    if (MeanTime >= 0 and MeanTime < 5):
        return 1
    elif (MeanTime >= 21 and MeanTime <= 24):
        return 1
    elif (MeanTime >= 5 and MeanTime < 13):
        return 2
    else :
        return 3
```

```python
OrderTime = np.zeros(206209)

for i in range(206209):
    if present[i] > 0:
        OrderTime[i] = ActiveTime(i+1)

OrderTime = pd.DataFrame({'Order Time' : OrderTime})
OrderTime
```

```python
sns.countplot(x = 'Order Time',data = OrderTime)
```

```python
sum(orders['days_since_prior_order'].isna())
```

```python
#For frequency of orders placed by a user, we can check the total orders in a fixed time duration by a user
#But the dataset doesn't contain information regarding Date of order placed so we will use Total Numbers of Orders
#a user as a close substitute

Frequency = present

Frequency = pd.DataFrame({'Order Frequency' : Frequency})

sns.countplot(x = 'Order Frequency',data = Frequency)
```

```python
#Unlike standard RMF segmentation, we also try to include other features which are significant in the dataset

sns.countplot(x = 'order_dow',data = orders)
```

```python
#Recency of orders placed by a user can be evaluated by average time in days after which the next order
#is placed by the user

DayLag = np.zeros(206209)

for i in range(206209):
    if present[i] > 0:
        DayLag[i] = orders.loc[orders['user_id'] == i+1]['days_since_prior_order'].mean()

DayLag = pd.DataFrame({'Order Recency' : DayLag})

sns.countplot(x = 'Order Recency',data = DayLag)
```

```python
DayLag
```

```python
#As the prices of the products orders are not listed in the dataset, so for the 'Monetary Segmentation' we decide
#through the average size of an order placed by a user. This makes us assume that all the products are nearly 
#equally priced

OrderSize = np.zeros(206209)

for i in range(316717):
    UserID = int(orders.loc[i]['user_id'])
    OrderID = int(orders.loc[i]['order_id'])
    OrderSize[UserID - 1] += sum(orderedProducts.loc[orderedProducts['order_id'] == OrderID]['add_to_cart_order'])

for i in range(206209):
    if present[i] != 0:
        OrderSize[i] //= present[i]

OrderSize = pd.DataFrame({'Order Size' : OrderSize})
```

```python
df = pd.concat([OrderSize,Frequency,OrderTime,DayLag],ignore_index = True,axis = 1)
df.rename(columns = {0 : 'Order Size',1 : 'Order Frequency',2 : 'Order Time',3 : 'Order Recency'},inplace = True)
df = df.loc[df['Order Frequency'] != 0]
df.reset_index(inplace = True)
df.head()
```

```python
df.drop('index',axis = 1,inplace = True)
```

```python
df
```

```python
#Normalizing the data

from sklearn.preprocessing import normalize

ND = normalize(df)
ND = pd.DataFrame(data = ND)
ND.describe()
```

```python
ND.to_csv(r'NormalizedData.csv',index = False)
```

```python
#Reducing the dimensions to 2D for visualization

from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

TwoDimData = pca.fit_transform(ND)

TwoDimData = pd.DataFrame(data = TwoDimData,columns = ('Dim1','Dim2'))

TwoDimData.shape
```

```python
plt.scatter(TwoDimData['Dim1'],TwoDimData['Dim2'])
plt.xlabel('Dim1')
plt.ylabel('Dim2')
```

```python
from sklearn.cluster import KMeans
```

```python
#We measure the performance of k-means for different values of k through distortion-score (intra-cluster distance)
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,12))

visualizer.fit(ND)        
visualizer.show()
```

```python
kmeans3 = KMeans(n_clusters = 3).fit(ND)
kmeans3.labels = kmeans3.labels_
pca_data3 = PCA(n_components = 2).fit_transform(ND)


pca_df3 = pd.DataFrame(data = pca_data3,columns = ['dim1','dim2'])
sns.scatterplot(x = pca_df3['dim1'],y = pca_df3['dim2'],hue = kmeans3.labels)
```

```python
kmeans4 = KMeans(n_clusters = 4).fit(ND)
kmeans4.labels = kmeans4.labels_
pca_data4 = PCA(n_components = 2).fit_transform(ND)


pca_df4 = pd.DataFrame(data = pca_data4,columns = ['dim1','dim2'])
sns.scatterplot(x = pca_df4['dim1'],y = pca_df4['dim2'],hue = kmeans4.labels)
```

```python
#We also use silhouette_score(inter - cluster distance) to determine the optimal k
from sklearn.metrics import silhouette_score

SS4 = silhouette_score(ND,kmeans4.labels)
SS3 = silhouette_score(ND,kmeans3.labels)
```
